```python
x = np.linspace(-5,5,1000)
for a in range(200,500):
    y = x**2 + a*x/100-10
    plt.plot( x,y )
plt.xlabel( 'linear values' )
plt.ylabel( 'function values' )
plt.title( 'Quadratic equation' )
```
