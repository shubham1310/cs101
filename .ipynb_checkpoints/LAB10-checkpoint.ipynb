# Images

```python
import matplotlib.pyplot as plt
import numpy as np
from imageio import imread 
```

```python
img=imread('q.jpg')
imgplot = plt.imshow(img)
```

```python
print(img[0][0])
print(img.shape)
```

```python
img[100:1000,100:1000,:]=0
imgplot = plt.imshow(img)
```

```python
img=imread('q.jpg')
i=100;j=200;h=1000
img[i:i+h,j:j+h]=img[i:i+h,j:j+h].mean()
imgplot = plt.imshow(img)
```

```python
img=imread('q.jpg')
h=200
left=100; top=100; width=1000; length=1000;
for i in np.arange(left,left+width,h):
    for j in np.arange(top,top+length,h):
        img[i:i+h,j:j+h]=(i+j)%255
imgplot = plt.imshow(img)
```
